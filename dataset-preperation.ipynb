```python
a=pd.read_csv('train_po_cdhit.aac.csv')
b=pd.read_csv('train_ne_cdhit.aac.csv')
a['class']=1
b['class']=0
df=pd.concat([a,b],axis=0)
df.to_csv('train-combined-cdhit-aac.csv',index=False)
```

```python
a=pd.read_csv('train_po_cdhit.atc.csv')
b=pd.read_csv('train_ne_cdhit.atc.csv')
df=pd.concat([a,b],axis=0)
df.to_csv('train-combined-cdhit-atc.csv',index=False)
```

```python
a=pd.read_csv('train_po_cdhit.btc.csv')
b=pd.read_csv('train_ne_cdhit.btc.csv')
df=pd.concat([a,b],axis=0)
df.to_csv('train-combined-cdhit-btc.csv',index=False)
```
